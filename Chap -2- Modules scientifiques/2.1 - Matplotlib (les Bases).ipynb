## Matplotlib : Graphiques de Base

### Pyplot
matplotlib.pyplot est une interface basée sur l'état de matplotlib. Il fournit un moyen de traçage implicite, semblable à MATLAB. Il ouvre également les figures sur votre écran et agit en tant que gestionnaire de l'interface graphique des figures.
link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html

pyplot est principalement destiné aux tracés interactifs et aux cas simples de génération de tracés programmatiques :

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Graphiques simples

```python
X = np.linspace(0, 2, 10)
y = X**2

plt.plot(X, y)
plt.show()
```

```python
plt.scatter(X, y)
plt.show()
```

### Styles Graphiques
Il existe beaucoup de styles a ajouter aux graphiques. Voici les plus importants a retenir :
- **c** : couleur de la ligne
- **lw** : epaisseur de la ligne (pour les graphiques plot)
- **ls** : style de la ligne (pour les graphiques plot)
- **size** : taille du point (pour les graphiques scatter)
- **marker** : style de points (pour les graphiques scatter)
- **alpha** : transparence du graphique



```python
plt.plot(X, y, c='red', lw=4, ls='--', alpha=0.8)
plt.show()
```

### Cycle de vie d'une figure
Pour créer des figures proprements, on doit suivre le cycle de vie suivant :
1. **plt.figure**(figsize())
2. **plt.plot()**
3. Extras (titre, axes, legendes)
4. **plt.show()**

```python
X = np.linspace(0, 5, 20)

plt.figure() # Création d'une figure
plt.plot(X, np.sin(X), label='sin(x)') # premiere courbe
plt.plot(X, np.cos(X), label='cos(x)') # deuxieme courbe
# Extra information
plt.title('figure 1') # titre
plt.xlabel('axe x') # axes
plt.ylabel('axe y') # axes
plt.legend() # legend

plt.savefig('figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

### Subplots
Les subplots sont un autre éléments a ajouter pour créer plusieurs graphiques sur une meme figure

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='blue')
```

### Méthode orientée objet

```python
fig, ax = plt.subplots(2, 1, sharex=True) # partage le meme axe pour les subplots
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```
